## Why this matters for data science

You will never receive a perfectly clean dataset in a professional setting. Every real-world dataset has missing values, unexpected types, outliers, and surprises. EDA is the process of discovering those problems before they become model failures or misleading conclusions. Skipping EDA is one of the most common mistakes junior data scientists make, and one of the most expensive.

## What's happening?

Exploratory Data Analysis is the practice of systematically examining a dataset before drawing any conclusions. The goal is not to answer the final question yet. The goal is to understand what the data contains, where it has problems, and what patterns are worth investigating.

A typical EDA covers:

| Question | What you are checking |
|----------|-----------------------|
| What is the shape of the data? | Rows, columns, data types |
| What is missing? | Null counts, missing patterns |
| What is the range of each variable? | Minimums, maximums, distributions |
| Are there outliers? | Extreme values that might be errors |
| Do any variables seem related? | Correlations, grouped patterns |

EDA is not linear. You start with broad questions and follow the data wherever it leads. A finding in one variable often raises a question about another.

## Real-world example: Titanic Dataset

The Titanic dataset records information about 891 passengers aboard the RMS Titanic, which sank on April 15, 1912. It is one of the most studied datasets in data science education because it is small enough to understand completely, messy enough to be realistic, and meaningful enough to stay interesting.

The question we will investigate across all 12 notebooks in this series is: **what factors influenced whether a passenger survived?**

Every notebook from here forward uses this dataset as the primary example. By the final notebook, you will have conducted a complete EDA from raw data to actionable insights.

Notice:
- **891 passengers** are recorded, representing roughly 40% of the 2,224 people aboard
- Only **342 passengers survived**, a survival rate of about 38%
- The data contains a mix of numeric columns (age, fare, number of siblings) and categorical columns (sex, passenger class, port of embarkation)

> **Discussion question:** Before looking at any numbers, what factors do you think influenced whether someone survived the Titanic? Write down 3 hypotheses.

```python
np.random.seed(42)

survival_counts = titanic["survived"].value_counts().sort_index()
labels = ["Did not survive", "Survived"]
colors = [PALETTE["secondary"], PALETTE["accent"]]

fig = go.Figure(
    data=[go.Bar(
        x=labels,
        y=survival_counts.values,
        marker_color=colors,
        text=survival_counts.values,
        textposition="outside",
        width=0.5,
    )],
    layout=base_layout(
        title="Titanic Survival Counts (n=891)",
        xaxis_title="Outcome",
        yaxis_title="Number of Passengers",
    ),
)
fig.update_layout(showlegend=False, yaxis=dict(range=[0, 620]))
fig.show()
```
