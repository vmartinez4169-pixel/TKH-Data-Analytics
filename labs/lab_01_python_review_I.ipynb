```python
# Exercise B1 — Tip percentage function
# Complete the function so it returns the tip as a percentage of the total bill.

def tip_percentage(tip, total):
    return tip / total * 100

# --- check ---
assert tip_percentage(3, 15) == 20.0,   "Hint: tip / total * 100 — check which argument is which"
assert tip_percentage(5, 25) == 20.0,   "Hint: tip / total * 100"
assert tip_percentage(4, 20) == 20.0,   "Hint: tip / total * 100"
print("B1 correct!")
```

```python
# Exercise B2 — Filter high tips with a for loop
# Complete the loop so high_tips contains every tip amount greater than $5.

high_tips = []
for tip in tips['tip']:
    if tip > 5:
        high_tips.append(tip)

# --- check ---
assert isinstance(high_tips, list), "high_tips should be a list"
assert len(high_tips) > 0, "high_tips should not be empty — check your condition"
assert all(t > 5 for t in high_tips), "All tips in high_tips should be greater than 5"
assert len(high_tips) == (tips['tip'] > 5).sum(), "You may have the wrong condition — use >"
print(f"B2 correct! Found {len(high_tips)} tips over $5.")
```

```python
# Exercise B3 — Dictionary of average tips by day
# Complete the loop so avg_tips maps each day name to the average tip for that day.

avg_tips = {}
for day in tips['day'].unique():
    avg_tips[day] = tips[tips['day'] == day]['tip'].mean()

# --- check ---
assert isinstance(avg_tips, dict), "avg_tips should be a dictionary"
assert len(avg_tips) == tips['day'].nunique(), "You should have one entry per unique day"
assert all(isinstance(k, str) for k in avg_tips.keys()), "Keys should be day name strings"
assert all(0 < v < 20 for v in avg_tips.values()), "Values should be reasonable tip averages"
print("B3 correct! Days found:", list(avg_tips.keys()))
```

## Section C — Guided Function

Write a function called `summarize_tips` that takes the tips DataFrame and returns
a dictionary with these four keys:

| Key | Value |
|-----|-------|
| `'total_bills'` | Total number of bills (rows) in the dataset |
| `'avg_tip'` | Average tip amount, rounded to 2 decimal places |
| `'best_day'` | The day with the highest average tip |
| `'tip_rate'` | Overall tip percentage (sum of tips / sum of total bills × 100), rounded to 2 decimal places |

**Guidance:** Work one key at a time. Build the dictionary, then return it.

```python
def summarize_tips(df):

    avg_by_day = {}

    for day in df['day'].unique():
        avg_by_day[day] = df[df['day'] == day]['tip'].mean()

    best_day = max(avg_by_day, key=avg_by_day.get)

    summary = {
        'total_bills': len(df),
        'avg_tip': round(df['tip'].mean(), 2),
        'best_day': best_day,
        'tip_rate': round((df['tip'].sum() / df['total_bill'].sum()) * 100, 2)
    }

    return summary


result = summarize_tips(tips)
result  # should display the dictionary
```

```python
# --- Section C checks ---
assert isinstance(result, dict),     "summarize_tips should return a dictionary, not a " + str(type(result).__name__)

assert set(result.keys()) == {'total_bills', 'avg_tip', 'best_day', 'tip_rate'},     "Your dictionary is missing keys or has extra ones. Required: total_bills, avg_tip, best_day, tip_rate"

assert result['total_bills'] == len(tips),     f"total_bills should be {len(tips)} (the number of rows), got {result.get('total_bills')}"

assert result['best_day'] in tips['day'].cat.categories.tolist(),     f"best_day should be a day in the dataset, got {result.get('best_day')!r}"

assert 0 < result['tip_rate'] < 100,     f"tip_rate should be a percentage between 0 and 100, got {result.get('tip_rate')}"

assert isinstance(result['avg_tip'], float),     "avg_tip should be a float"

print("All Section C checks passed!")
```
